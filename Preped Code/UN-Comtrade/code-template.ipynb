
# üåê UN Comtrade Data Overview

The **UN Comtrade** (United Nations Commodity Trade Statistics Database) provides detailed data on international trade in **goods** (not services).

---

## üì¶ Trade Data Content

You can find **imports and exports** between countries, including:

- **What** was traded (commodity)
- **Who** traded it (reporting and partner countries)
- **When** it was traded (year/month)
- **How much** was traded (value and quantity)

---

## üßæ Key Data Fields

| Field               | Description                                                 |
|--------------------|-------------------------------------------------------------|
| **Reporter**        | The country providing the trade data                        |
| **Partner**         | The trading partner country                                 |
| **Year**            | Year of the trade                                           |
| **Trade Flow**      | Import, Export, or Re-export                                |
| **Commodity Code**  | Based on the Harmonized System (HS) or SITC classification  |
| **Commodity Description** | Name of the traded product                        |
| **Trade Value**     | Value in USD                                                |
| **Quantity**        | Physical quantity (e.g. kilograms, liters, units)           |
| **Unit**            | Unit of measure                                             |
| **Customs Code**    | (Sometimes) additional codes for specific regulations       |

---

## üìò Classification Systems

- **HS (Harmonized System)** ‚Äì Most common (e.g., HS 2012, HS 2017)
- **SITC (Standard International Trade Classification)** ‚Äì For economic analysis
- **BEC (Broad Economic Categories)** ‚Äì Groups goods by economic function
- **EBOPS** ‚Äì For trade in services (separate dataset)

---

## üåç Geographic Scope

- Data from **over 200 countries** and territories
- Reported **annually** or **monthly**, depending on the country

---

## üîç Use Cases

- Market research (e.g., avocado exports to Japan)
- Trade policy analysis
- Supply chain and logistics insights
- Monitoring sanctions or trade imbalances

---

## üéØ Step 3: Set Your Search Filters

You'll be presented with a form. Set the following:

- **Year(s):** Choose the desired year(s)
- **Commodity Code:** Use HS/SITC codes or search by keyword
- **Trade Flow:** Import, Export, or Both
- **Classification System:** e.g., HS 2017

Click **‚ÄúPreview‚Äù** to check your selection.
