# üìä Types of Data on the FRED (Federal Reserve Economic Data) Site

[FRED](https://fred.stlouisfed.org) is a vast repository of economic data curated by the Federal Reserve Bank of St. Louis. Here's what you can find:

---

## 1. Macroeconomic Indicators
- **GDP (Gross Domestic Product)**
- **CPI (Consumer Price Index)**
- **PCE (Personal Consumption Expenditures)**
- **GNP**, **National Income**, **Savings Rates**

---

## 2. Labor Market Data
- **Unemployment Rate**
- **Labor Force Participation Rate**
- **Employment-Population Ratio**
- **Job Openings and Labor Turnover Survey (JOLTS)**
- **Nonfarm Payrolls**
- **Wages and Earnings**

---

## 3. Inflation & Prices
- **Consumer Price Index (CPI)**
- **Producer Price Index (PPI)**
- **Inflation Expectations**
- **Core vs. Headline Inflation**

---

## 4. Monetary & Financial Data
- **Federal Funds Rate**
- **Interest Rates** (Treasury yields, mortgage rates, corporate bonds, etc.)
- **Money Supply (M1, M2)**
- **Bank Reserves and Lending**
- **Yield Curve Data**

---

## 5. Housing & Real Estate
- **Housing Starts**
- **New and Existing Home Sales**
- **Home Price Indices** (e.g., Case-Shiller Index)
- **Mortgage Rates**
- **Delinquency and Foreclosure Rates**

---

## 6. International Economic Data
- **GDP and Inflation for Other Countries**
- **Exchange Rates**
- **Global Trade Balances**
- **Country-Specific Indicators** (e.g., China, EU, Japan)

---

## 7. Trade & Industry
- **Import and Export Data**
- **Industrial Production**
- **Capacity Utilization**
- **Retail and Wholesale Trade Figures**

---

## 8. Government Finance
- **Federal Government Debt**
- **Expenditures and Revenues**
- **Budget Deficits**
- **State and Local Government Finances**

---

## 9. Forecasts and Expectations
- Data from:
  - **Survey of Professional Forecasters**
  - **University of Michigan Consumer Sentiment**


# üì• How to Download Data from FRED (Federal Reserve Economic Data)

Follow these steps to find and download datasets from [FRED](https://fred.stlouisfed.org):

---

## ‚úÖ Step 1: Visit FRED

Go to: [https://fred.stlouisfed.org](https://fred.stlouisfed.org)

---

## üîé Step 2: Search for a Data Series

- Use the search bar at the top to find data by:
  - Keyword (e.g. "unemployment rate", "GDP", "CPI")
  - Series ID (e.g. `UNRATE`, `GDP`, `CPIAUCSL`)
- Browse by **Category**, **Source**, or **Release Calendar** if you're not sure what you're looking for.

---

## üìä Step 3: Select a Series

- Click on the series name (e.g. *Unemployment Rate*).
- You‚Äôll be taken to a page with an interactive graph and summary info.

---

## üì§ Step 4: Download the Data

1. On the series page, click the **"Download"** button (top-right of the graph).
2. Choose:
   - **File format:** CSV, XLSX, or XML
   - **Date range** (optional)
3. Click **‚ÄúDownload‚Äù** and save the file to your computer.

---

## üõ† Optional: Adjust Graph Settings Before Download

Before downloading, you can customize:
- **Units** (e.g. percent change, index value)
- **Frequency** (monthly, quarterly, annually)
- **Aggregation method** (average, sum, end of period)
